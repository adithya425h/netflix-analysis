```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
df = pd.read_csv('NetflixDataset.csv')

# First look at the data
print("First 5 rows:")
display(df.head())

print("\nColumn info:")
df.info()

print("\nMissing values per column:")
print(df.isnull().sum())

# Visual 1: Count of content type (Movie vs TV Show)
plt.figure(figsize=(6, 4))
sns.countplot(data=df, x='type', palette='Set2')
plt.title("Count of Content Type")
plt.show()

# Visual 2: Top 10 countries by number of shows
plt.figure(figsize=(10, 5))
df['country'].value_counts().head(10).plot(kind='bar', color='coral')
plt.title("Top 10 Countries by Netflix Content")
plt.ylabel("Number of Shows")
plt.xticks(rotation=45)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
df = pd.read_csv('NetflixDataset.csv')

# First look at the data
print("First 5 rows:")
display(df.head())

print("\nColumn info:")
df.info()

print("\nMissing values per column:")
print(df.isnull().sum())

# Visual 1: Count of content type (Movie vs TV Show)
plt.figure(figsize=(6, 4))
sns.countplot(data=df, x='type', palette='Set2')
plt.title("Count of Content Type")
plt.show()

# Visual 2: Top 10 countries by number of shows
plt.figure(figsize=(10, 5))
df['country'].value_counts().head(10).plot(kind='bar', color='coral')
plt.title("Top 10 Countries by Netflix Content")
plt.ylabel("Number of Shows")
plt.xticks(rotation=45)
plt.show()
```

```python
df_2020 = df[df['release_year'] == 2020]
print(f"Total shows released in 2020: {len(df_2020)}")
display(df_2020.head())
```

```python
drama_shows = df[df['listed_in'].str.contains("Drama", na=False)]
print(f"Total Drama shows: {len(drama_shows)}")
display(drama_shows.head())
```

```python
plt.figure(figsize=(10, 5))
df['release_year'].value_counts().sort_index().plot(kind='line', color='skyblue', marker='o')
plt.title("Netflix Content Released Over Time")
plt.xlabel("Release Year")
plt.ylabel("Number of Titles")
plt.grid(True)
plt.show()
```

```python
top_directors = df['director'].dropna().value_counts().head(10)

plt.figure(figsize=(8, 5))
top_directors.plot(kind='barh', color='salmon')
plt.title("Top 10 Directors on Netflix")
plt.xlabel("Number of Titles")
plt.gca().invert_yaxis()
plt.show()
```

```python
# Split actor lists and count occurrences
from collections import Counter

all_actors = df['cast'].dropna().str.split(', ')
flat_actor_list = [actor for sublist in all_actors for actor in sublist]
top_actors = Counter(flat_actor_list).most_common(10)

# Convert to DataFrame for plotting
top_actors_df = pd.DataFrame(top_actors, columns=['Actor', 'Count'])

plt.figure(figsize=(8, 5))
sns.barplot(data=top_actors_df, y='Actor', x='Count', palette='magma')
plt.title("Top 10 Actors on Netflix")
plt.xlabel("Number of Titles")
plt.show()
```
